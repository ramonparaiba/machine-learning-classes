```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets-colab/w3schools/data.csv')
```

```python
#Correlação entre as colunas
#o valor das correlações varia de -1 a 1. 
#quanto mais próximos dos extremos, maior a correlação
#quanto mais próximo de zero, menor a correlação
#uma boa correlação começa a partir de 0.6 (ou -0.6)(pesquisar melhor os porques)
df.corr()
#Análise preliminar
#Duração e calorias, tiveram uma alta correlação (0.922717)
#pode-se inferir que, quanto maior a duração do exercício, 
#maior o número de calorias queimadas
```

- Análise preliminar
***
- Boa Correlação - Duração e Calorias, tiveram uma boa correlação (0.922717) 
 - entende-se que quanto maior a duração do exercício maior o número de calorias queimadas
- Correlação ruim - Duração e MaxPulse, tiveram uma correlação baixa (0.009403). Ou seja, não dá pra prever o pulso máximo, apenas observando a duração do exercício.
