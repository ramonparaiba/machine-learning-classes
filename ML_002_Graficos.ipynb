```python
#importação do numpy e do  Matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

#Gráficos de dispersão
- Um gráfico de dispersão é um diagrama em que cada valor no conjunto de dados é representado por um ponto.

```python
#Para renderizar um gáfico de dispersão, o Matplotlib precisa de dois arrays do mesmo tamanho (para X e Y)
ageCar = [5,7,8,7,2,17,2,9,4,11,12,9,6]
speedCar = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plt.scatter(ageCar, speedCar)
plt.show()
```

#distribuição aleatória de dados

```python
#Criação de duas bases aleatórias
x = np.random.normal(5.0, 1.0, 1000)
y = np.random.normal(10.0, 2.0, 1000)
```

```python
plt.scatter(x, y)
plt.show()

#Conforme estabelecido na geração dos dados, os pontos estão concentrados em 5(eixo X) e em 10(eixo Y)
```
